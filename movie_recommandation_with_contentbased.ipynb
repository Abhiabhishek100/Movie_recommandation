```python

import pandas as pd

# Load CSV files into dataframes
df1 = pd.read_csv('movies.csv')
df2 = pd.read_csv('ratings.csv')

# Merge dataframes based on a common column
merged_df = pd.merge(df1, df2, on='movieId')

# Write the merged dataframe to a new CSV file
merged_df.to_csv('merged_file.csv', index=False)
```

```python
import pandas as pd

# Assuming df is your dataframe
# df = pd.DataFrame(...)

# Specify the column(s) you want to check for duplicates
column_to_check = 'column_name'

# Remove rows with similar values in the specified column
df_no_duplicates = df1.drop_duplicates(subset=["movieId"])

# Optionally, you can reset the index if you want a clean index after removing duplicates
df_no_duplicates = df_no_duplicates.reset_index(drop=True)
```

```python
df2=df_no_duplicates
```

```python
df2
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity




tfidf_vectorizer = TfidfVectorizer()

tfidf_matrix = tfidf_vectorizer.fit_transform(df2['genres'])

# Calculate cosine similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Create a DataFrame to store cosine similarities
cosine_sim_df = pd.DataFrame(cosine_sim, columns=df2.index, index=df2.index)

print("Cosine Similarity Matrix:")
print(cosine_sim_df)
```

```python
sim=cosine_sim_df
```

```python
df5=sim[4]
```

```python
sorted_data = df5.sort_values(ascending=False)
print(sorted_data)
```

```python
df2.iloc[sorted_data.index]
```

```python
def movie_rec(Title):
    index=df2[df2["title"]==Title]["movieId"].iloc[0]
    # print(index)
    df5=sim[index-1]
    # print(df5)
    sorted_data = df5.sort_values(ascending=False)
    # print(sorted_data)
    return df2.iloc[sorted_data.index]["title"]
    
    
    
    
```
